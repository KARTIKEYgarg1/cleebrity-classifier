```python
def get_cropped_images_if2_eyes(imagepath):
    t1=cv2.imread(imagepath)
    t2=cv2.cvtColor(t1,cv2.COLOR_BGR2GRAY)
    face_cascade=cv2.CascadeClassifier("./opencv/harcascade_frontalface_default.xml")
    eye_cascade=cv2.CascadeClassifier("./opencv/harcascade_eye.xml")
    #detecting number of faces
    faces=face_cascade.detectMultiScale(t2,1.3,5)
    #selecting faces one by one
    #here i returned on the first face only, change to len(face>0) for multiple faces
    for (x,y,w,h) in faces:
        t3=cv2.rectangle(t1,(x,y),(x+w,y+h),(255,0,0),2)
        t2=cv2.cvtColor(t1,cv2.COLOR_BGR2GRAY)
        eyes=eye_cascade.detectMultiScale(t2)
    #selecting the 2 eyes,rest are anamoly which are disgarded like crowd eyes
#         for (ex,ey,ew,eh) in eyes:
#             t3=cv2.rectangle(t3,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        t3=t3[y:y+h,x:x+w]#cropping only face
        if len(eyes)>1 and len(faces)>0:
            return t3
```

```python
import os
img_dirs=[]
for entry in os.scandir("./dataset"):
    if entry.is_dir():
        img_dirs.append(entry.path)
img_dirs
```

```python
import shutil
if os.path.exists("./dataset/cropped/"):
    shutil.rmtree("./dataset/cropped/")
#removing the old directory and making a new directory
os.mkdir("./dataset/cropped/")
```

```python
cr_img_dir=[]
cel_files_name_dict={}
for img_dir in img_dirs:
    count=0
    cel_name=img_dir.split("\\")[-1]
    cel_files_name_dict[cel_name]=[]
#     print(cel_name)
    for entry in os.scandir(img_dir):
        cropped_face_with_eyes=get_cropped_images_if2_eyes(entry.path)
        if cropped_face_with_eyes is not None:
            crFolder="./dataset/cropped/"+cel_name
            if not os.path.exists(crFolder):
                os.mkdir(crFolder)
            crNewFileName=crFolder+"/"+cel_name+str(count)+".jpeg"
            cel_files_name_dict[cel_name].append(crNewFileName)
            cv2.imwrite(crNewFileName,cropped_face_with_eyes)
            count=count+1
#             plt.imshow(cropped_face_with_eyes)
```

```python
import numpy as np
import pywt
import cv2
def w2d(img,mode="haar",level=1):
    imArray=img
    imArray=cv2.cvtColor(imArray,cv2.COLOR_BGR2GRAY)
    imArray=np.float32(imArray)
    imArray /= 255
    coeff=pywt.wavedec2(imArray,mode,level=level)
    coeff_n=list(coeff)
    coeff_n[0]*=0
    imArray_n=pywt.waverec2(coeff_n,mode)
    imArray_n*=255
    imArray_n=np.uint8(imArray_n)
    return imArray_n
```

```python
imhar=w2d(cv2.imread(cel_files_name_dict['elon_musk'][3]),'db1',5)
plt.imshow(imhar,cmap='gray')
# plt.imshow(cv2.imread(cel_files_name_dict['elon_musk'][3]))
```

```python
class_dict={}
count=0
for cel_name in cel_files_name_dict.keys():
    class_dict[cel_name]=count
    count+=1
class_dict
```

```python

X,Y=[],[]
for cel_name,trainFiles in cel_files_name_dict.items():
    for training_image in trainFiles:
        img=cv2.imread(training_image)
        if img is None:
            continue
        scalled_raw=cv2.resize(img,(32,32))
        img_har=w2d(img,'db1',5)
        scalled_raw_har=cv2.resize(img_har,(32,32))
        combinedImg=np.vstack((scalled_raw.reshape(32*32*3,1),scalled_raw_har.reshape(32*32,1)))
        X.append(combinedImg)
        Y.append(class_dict[cel_name])
print(Y)
```

```python
X=np.array(X).reshape(len(X),4096).astype(float)
X.shape
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=3)
pipe=Pipeline([('scaler',StandardScaler()),('svc',SVC(kernel='linear',C=1))])
# pipe=Pipeline([('scaler',StandardScaler()),('random_forest',RandomForestClassifier(n_estimators=1))])
# pipe=Pipeline([('scaler',StandardScaler()),('logistic_regression',LogisticRegression(C=5))])
pipe.fit(X_train,Y_train)
pipe.score(X_test,Y_test)
```

```python
len(X_test)
```

```python
print(classification_report(Y_test,pipe.predict(X_test)))
```

```python
model_params={
    'svm':{
        'model':svm.SVC(gamma='auto',probability=True),
        'params':{
            'svc__C':[1,10],
            'svc__kernel':['rbf','linear']
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'logisticregression__C':[1,10]
        }
    }
}
```

```python
scores=[]
best_estimator={}
import pandas as pd
for algo,mp in model_params.items():
    pipe=make_pipeline(StandardScaler(),mp['model'])
    clf=GridSearchCV(pipe,mp['params'],cv=5,return_train_score=False)
    clf.fit(X_train,Y_train)
    scores.append({'model':algo,'best_score':clf.best_score_,'best_params':clf.best_params_})
    best_estimator[algo]=clf.best_estimator_
```

```python
df=pd.DataFrame(scores)
df
```

```python
best_clf=best_estimator['svm']
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,best_clf.predict(X_test))
cm
```
